```python
import sympy as sp
import numpy as np

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

from tabulate import tabulate
```

```python
def f(x):
  return np.sin(np.pi*x)
```

```python
def tridiagonal(n, a, b):
  return sp.Matrix([
      [a if i == j else 0 for j in range (n)]
      for i in range(n)
  ])
```

```python
r=1
```

```python
A = tridiagonal(9, 2-2*r, r)
```

```python
A
```

```python
u0 = sp.Matrix([f(0.1*(i+1)) for i in range(9)])
```

```python
u1=u0
```

```python
def onda (A, u0, u1, j):
  u0 = sp.Matrix(u0)
  u1 = sp.Matrix(u1)
  S = [u0, u1]

  for n in range (2, j):
    sol = A * S[n-1] - S[n-2]
    S.append(sol)

  return sp.Matrix([s.T for s in S])
```

```python
j=50
```

```python
U=onda(A, u0, u1, j)
```

```python
U_np = np.array(U.tolist(), dtype=float)
```

```python
x = np.linspace(0, 1, 9)
y = np.linspace(0, 0.5, j)

X, Y = np.meshgrid(x, y)
```

```python
# Mapa de colores
plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, U_np, levels=50, cmap=cm.viridis)
plt.colorbar(contour)
plt.title('Solución numérica de un problema de Dirichlet')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```
