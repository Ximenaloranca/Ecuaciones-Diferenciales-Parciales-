#<font color=pink> El presente trabajo es una grafica de la funcion parabólica y=x²


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores de x
x = np.linspace(-100, 100, 400)
y = x**2  # Función y = x^2

# Crea gráfica
plt.figure(figsize=(10, 8))
plt.plot(x, y, label=r'$y = x^2$', color='pink')

# Diseño de la gráfica
plt.axhline(0, color='purple', linewidth=0.5)
plt.axvline(0, color='purple', linewidth=0.5)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la funcion')

# Mostrar la gráfica
plt.show()
```
