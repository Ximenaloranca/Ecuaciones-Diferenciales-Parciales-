```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
from collections import defaultdict
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch

# === Definición del proyecto (hacer un pastel) ===
actividades = {
    "A": {"descripcion": "Buscar receta y lista de ingredientes", "duracion": 10},
    "B": {"descripcion": "Comprar ingredientes", "duracion": 40},
    "C": {"descripcion": "Preparar utensilios y precalentar horno", "duracion": 15},
    "D": {"descripcion": "Lavar y pesar ingredientes", "duracion": 15},
    "E": {"descripcion": "Mezclar ingredientes secos", "duracion": 10},
    "F": {"descripcion": "Batir ingredientes líquidos", "duracion": 15},
    "G": {"descripcion": "Integrar mezcla líquida con secos", "duracion": 20},
    "H": {"descripcion": "Engrasar y enharinar molde", "duracion": 5},
    "I": {"descripcion": "Verter mezcla en el molde", "duracion": 5},
    "J": {"descripcion": "Hornear el pastel", "duracion": 45},
    "K": {"descripcion": "Preparar relleno o cobertura", "duracion": 30},
    "L": {"descripcion": "Dejar enfriar pastel", "duracion": 20},
    "M": {"descripcion": "Decorar pastel", "duracion": 25},
    "N": {"descripcion": "Presentar y servir", "duracion": 5},
}

precedencias = [
    ("A", "B"),
    ("B", "D"),
    ("C", "H"),
    ("D", "E"),
    ("D", "F"),
    ("E", "G"),
    ("F", "G"),
    ("G", "I"),
    ("H", "I"),
    ("I", "J"),
    ("J", "L"),
    ("K", "M"),
    ("L", "M"),
    ("M", "N"),
]

# === Construir grafo ===
G = nx.DiGraph()
for k, v in actividades.items():
    G.add_node(k, duracion=v["duracion"], descripcion=v["descripcion"])
G.add_edges_from(precedencias)

# === Calcular ruta crítica sumando DURACIONES de nodos ===
topo = list(nx.topological_sort(G))

# dp[node] = duración máxima acumulada hasta node (incluye node)
dp = {}
prev = {}  # para reconstruir camino
for node in topo:
    preds = list(G.predecessors(node))
    if not preds:
        dp[node] = actividades[node]["duracion"]
        prev[node] = None
    else:
        best_pred = max(preds, key=lambda p: dp[p])
        dp[node] = dp[best_pred] + actividades[node]["duracion"]
        prev[node] = best_pred

# nodo final con mayor dp
end_node = max(dp.items(), key=lambda x: x[1])[0]
duracion_total = dp[end_node]

# reconstruir ruta crítica
crit_path = []
cur = end_node
while cur is not None:
    crit_path.append(cur)
    cur = prev[cur]
crit_path = list(reversed(crit_path))

# === Imprimir tabla ordenada por flujo lógico (topológico) ===
tabla = []
for node in topo:
    preds = [p for p, q in precedencias if q == node]
    tabla.append([node, actividades[node]["descripcion"], actividades[node]["duracion"],
                  ", ".join(preds) if preds else "-"])

df = pd.DataFrame(tabla, columns=["ID", "Descripción", "Duración (min)", "Precedencias"])
print("\nTabla de actividades (orden del flujo):\n")
print(df.to_string(index=False))

print("\nRuta crítica:", " -> ".join(crit_path))
print("Duración total (suma nodos en ruta crítica):", duracion_total, "minutos\n")

# === CALCULAR NIVELES (para layout horizontal, niveles = distancia en aristas desde fuentes) ===
levels = {}
for node in topo:
    preds = list(G.predecessors(node))
    if not preds:
        levels[node] = 0
    else:
        levels[node] = max(levels[p] + 1 for p in preds)

# agrupar nodos por nivel
nodes_by_level = defaultdict(list)
for n, lvl in levels.items():
    nodes_by_level[lvl].append(n)
max_level = max(levels.values())

# posiciones: x = nivel * x_space; y distribuidos verticalmente por nivel
x_space = 4.0
v_space = 1.6
box_w = 2.6
box_h = 0.9

pos = {}
for lvl in range(max_level + 1):
    nodes = nodes_by_level.get(lvl, [])
    n = len(nodes)
    if n == 0:
        continue
    # centrar verticalmente: indices 0..n-1 -> y from +(n-1)/2 to -(n-1)/2
    for i, node in enumerate(nodes):
        y = ( (n - 1) / 2.0 - i ) * v_space  # top to bottom
        x = lvl * x_space
        pos[node] = (x, y)

# determinar tamaño figura según niveles y cantidad máxima de nodos en un nivel
max_nodes_in_level = max(len(lst) for lst in nodes_by_level.values())
fig_w = max(10, (max_level + 1) * 2.2)
fig_h = max(5, max_nodes_in_level * 1.0)

# === DIBUJAR DIAGRAMA AON RECTO ===
fig, ax = plt.subplots(figsize=(fig_w, fig_h))
ax.set_aspect('equal')

# dibujar cajas y etiquetas
crit_edges = set(zip(crit_path, crit_path[1:]))

for node in topo:
    x, y = pos[node]
    dur = actividades[node]["duracion"]
    desc = actividades[node]["descripcion"]
    # color nodos en ruta critica
    face = "#ffb3b3" if node in crit_path else "#cfe2f3"
    # caja redondeada
    box = FancyBboxPatch((x - box_w/2, y - box_h/2),
                         box_w, box_h,
                         boxstyle="round,pad=0.3",
                         linewidth=1, edgecolor="black", facecolor=face)
    ax.add_patch(box)
    # etiqueta compacta: ID + descripción corta + duración
    short_desc = desc if len(desc) <= 28 else desc[:28] + "..."
    label = f"{node}: {short_desc}\n{dur} min"
    ax.text(x, y, label, ha='center', va='center', fontsize=8)

# dibujar flechas (edges)
for u, v in G.edges():
    x1, y1 = pos[u]
    x2, y2 = pos[v]
    start = (x1 + box_w/2, y1)
    end = (x2 - box_w/2, y2)
    is_crit = (u, v) in crit_edges
    arrow = FancyArrowPatch(start, end,
                            connectionstyle="arc3,rad=0.0",
                            arrowstyle='-|>',
                            mutation_scale=20,
                            linewidth=2 if is_crit else 1,
                            color='red' if is_crit else 'k')
    ax.add_patch(arrow)

# ajustar límites
xs = [p[0] for p in pos.values()]
ys = [p[1] for p in pos.values()]
pad_x = box_w
pad_y = box_h + 0.5
ax.set_xlim(min(xs) - pad_x, max(xs) + pad_x)
ax.set_ylim(min(ys) - pad_y, max(ys) + pad_y)

ax.axis('off')
plt.title("Red de Proyecto (AON) - Hacer un pastel", fontsize=12)
plt.tight_layout()
plt.show()
```
