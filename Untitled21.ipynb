```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

from tabulate import tabulate
```

```python
def f(x):
  return 1-4*(x-0.5)**2
```

```python
x=np.linspace(0,1,10)
plt.plot(x,f(x))
```

```python
def tridiagonal(n, a, b):
  return sp.Matrix([
      [a if i == j else b if abs (i-j) == 1 else 0 for j in range (n)]
      for i in range(n)
  ])
```

```python
r=1
```

```python
A = tridiagonal (9, 2 + 2*r, -r)
```

```python
A
```

```python
B = tridiagonal (9, 2 - 2*r, r)
```

```python
B
```

```python
b = sp.Matrix([f(0.1*(i+1)) for i in range (9)])
```

```python
b
```

```python
def crank_nicolson(A,B,b,j):
  b_1 = B*b
  S = []

  for i in range(j):
    sol = A.LUsolve(b_1)
    S.append(sol.T)
    b_1 = B*sol

  return sp.Matrix(S)
```

```python
j = 50
```

```python
U = crank_nicolson(A,B,b,j)
```

```python
data = [list(U.row(i)) for i in range(5)]

print(tabulate(data, tablefmt="grid"))
```

```python
U_np = np.array(U.tolist(), dtype =float)
```

```python
x = np.linspace(0,1,9)
y = np.linspace (0,0.5,j)
X, Y = np.meshgrid(x,y)
```

```python
plt.figure(figsize=(8,6))
contour = plt.contourf(X,Y, U_np, levels=25, cmap=cm.viridis)
plt.colorbar(contour)
plt.title('Solucion')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```
