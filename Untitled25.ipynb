```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x') # introducimos una variable
u = sp.Function('u') # introducimos una funcion
```

```python
ode = u(x).diff(x, x) + u(x) + x
sp.dsolve(ode, u(x))
```

```python
sol = sp.dsolve(ode, ics={u(0): 0, u(1): 0}) # Es un diccionario, tiene llaves y valor para las condiciones iniciales
```

```python
sol
```

```python
u_expr = sol.rhs


u_num = sp.lambdify(x, u_expr, "numpy")


xx = np.linspace(0, 1, 100)
yy = u_num(xx)

plt.figure(figsize=(8,5))
plt.plot(xx, yy,label="Solucion $u(x)$, linewidth=2")
plt.title("Solucion del problema de contorno\n$u' ' + u = -x$, $u(0)=u(1)=0$")
plt.xlabel("x")
plt.ylabel("u(x)")
plt.grid(True)
plt.legend()
plt.show()
```

ahora con las funciones de prueba

```python
a, b = sp.symbols('a b')

# funcion de prueba

u_n = a*x*(1-x) + b*x**2*(1-x)
```

```python
u_n
```

```python
u_n_2 = sp.diff(u_n, x, 2)


F = u_n_2 + u_n
```

```python
F
```

```python
# puntos de colocacion

points = [sp.Rational(1,4), sp.Rational(1,2)]
```

```python
points
```

```python
eqs = [sp.Eq(F.subs(x,p), -p) for p in points]

#Resolver para a y b
sol_1=sp.solve(eqs, [a, b])
```

```python
for eq in eqs:
  print(eq)
```

```python
print("Sistema de ecuaciones simbolico: ")
for eq in eqs:
  sp.pprint(eq)
```

```python
sol_1
```

```python
u_approx = u_n.subs(sol_1)
```

```python
u_approx
```

```python
u_approx_fn = sp.lambdify(x, u_approx, "numpy" )
```

```python
# solucion analitica y numerica
xx = np.linspace(0, 1, 100)
plt.figure(figsize=(9,5))
plt.plot(xx, u_approx_fn(xx), label="aaproximacion por colocacion", linewidth=2)
plt.plot(xx, yy, "--", label="solucion exacta", linewidth=2)

# puntos de colocacion
plt.scatter([ ])
```
